2. Remove samples whose permeability values are outliers. Show how the outliers were
determined and report the number of removed samples.  

Removed 26 outlier samples in In situ Klinkenberg Permeability (mD)=.

```python
import numpy as np

# Compute IQR
Q1 = df_cleaned["In situ Klinkenberg Permeability (mD)="].quantile(0.25)
Q3 = df_cleaned["In situ Klinkenberg Permeability (mD)="].quantile(0.75)
IQR = Q3 - Q1

# Define outlier thresholds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers
df_no_outliers = df_cleaned[(df_cleaned["In situ Klinkenberg Permeability (mD)="] >= lower_bound) &
                            (df_cleaned["In situ Klinkenberg Permeability (mD)="] <= upper_bound)]

# Report number of outliers removed
num_outliers_removed = len(df_cleaned) - len(df_no_outliers)
print(f"Removed {num_outliers_removed} outlier samples in In situ Klinkenberg Permeability (mD)=.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a figure
plt.figure(figsize=(8, 6))

# Boxplot before removing outliers
sns.boxplot(data=df_cleaned, y="In situ Klinkenberg Permeability (mD)=", color="lightblue")

# Title and labels
plt.title("Boxplot of In-Situ Klinkenberg Permeability (Before Outlier Removal)")
plt.ylabel("Permeability (mD)")
plt.show()
```

```python
# Create a figure
plt.figure(figsize=(12, 5))

# Histogram before removing outliers
plt.subplot(1, 2, 1)
sns.histplot(df_cleaned["In situ Klinkenberg Permeability (mD)="], bins=20, kde=True, color="blue")
plt.title("Permeability Distribution (Before Outlier Removal)")
plt.xlabel("Permeability (mD)")

# Histogram after removing outliers
plt.subplot(1, 2, 2)
sns.histplot(df_no_outliers["In situ Klinkenberg Permeability (mD)="], bins=20, kde=True, color="green")
plt.title("Permeability Distribution (After Outlier Removal)")
plt.xlabel("Permeability (mD)")

plt.tight_layout()
plt.show()
```

3. Show which feature-engineering methods help you be er analyze the data. Elaborate your
answer and provide relevant plots.

Log transformation (Since data is skewed)

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Apply log transformation
df_no_outliers["Log Permeability"] = np.log(df_no_outliers["In situ Klinkenberg Permeability (mD)="])
df_no_outliers["Log Porosity"] = np.log(df_no_outliers["Routine Porosity ="])

# Histogram plots before and after transformation
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sns.histplot(df_no_outliers["In situ Klinkenberg Permeability (mD)="], bins=20, ax=axes[0], kde=True)
axes[0].set_title("Original Permeability")

sns.histplot(df_no_outliers["Log Permeability"], bins=20, ax=axes[1], kde=True)
axes[1].set_title("Log Transformed Permeability")

plt.show()
```

Scatter Plot for Sample Depth, Log Permeability, Log Porosity

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Ensure correct column names
df_no_outliers = df_no_outliers.rename(columns={
    "Sample Depth ft (Plug Letter) =": "Sample Depth",
    "In situ Klinkenberg Permeability (mD)=": "Permeability",
    "Routine Porosity =": "Porosity"
})

# Create the crossplots
plt.figure(figsize=(15, 5))

# Scatter Plot: Sample Depth vs Permeability
plt.subplot(1, 3, 1)
sns.scatterplot(x='Sample Depth', y='Permeability', data=df_no_outliers)
plt.title('Sample Depth vs. Permeability')

# Scatter Plot: Sample Depth vs Porosity
plt.subplot(1, 3, 2)
sns.scatterplot(x='Sample Depth', y='Porosity', data=df_no_outliers)
plt.title('Sample Depth vs. Porosity')

# Scatter Plot: Permeability vs Porosity
plt.subplot(1, 3, 3)
sns.scatterplot(x='Permeability', y='Porosity', data=df_no_outliers)
plt.title('Permeability vs. Porosity')

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```
